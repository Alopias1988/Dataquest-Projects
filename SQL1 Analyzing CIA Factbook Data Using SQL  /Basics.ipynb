```python
%%sql
SELECT 
    MIN(population) min_pop, 
    MAX(population) max_pop, 
    MIN(population_growth) min_pop_growth, 
    MAX(population_growth) max_pop_growth
FROM facts;
```

```python
%%sql
SELECT *
FROM facts
WHERE population == (SELECT MAX(population) FROM facts);
```

```python
%%sql
SELECT *
FROM facts
WHERE population == (SELECT MIN(population) FROM facts);
```

```python
%%sql
SELECT AVG(population) average_pop, AVg(area) average_area
FROM facts
WHERE name <> 'World';
```

```python
%%sql
SELECT *
FROM facts
WHERE population > (SELECT AVG(population) FROM facts) and area > (SELECT AVG(area) FROM facts);
```

```python
# Which countries have the highest ratios of water to land? 
%%sql
SELECT name, area_land, area_water, CAST((area_water) as Float) / CAST((area_land) as Float) ratio_water_to_land FROM facts
ORDER BY ratio_water_to_land DESC
LIMIT 5
```

Which countries have more water than land?

```python
%%sql
SELECT * 
FROM facts
WHERE area_water > area_land;
```

Which countries will add the most people to their population next year?

```python
%%sql
SELECT name, population_growth*population new_population
FROM facts
ORDER BY new_population DESC
LIMIT 5
```

Which countries have a higher death rate than birth rate?

```python
%%sql
SELECT name, birth_rate, death_rate
FROM facts
ORDER BY death_rate > birth_rate DESC;
```
